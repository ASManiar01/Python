```python
'''Create a program that will call a function to generate KE billing details
on the basis of given input.
The function will only call other function for any processing and only print return 
values from calling function'''

def electricity(monthly_units): # function for electric charge calculation
    if monthly_units <= 100:
        elctric_charges = monthly_units * 9.43
        return elctric_charges
    elif monthly_units > 100 and monthly_units <= 200:
        elctric_charges = monthly_units * 10.29
        return elctric_charges
    elif monthly_units > 200 and monthly_units <= 300:
        elctric_charges = monthly_units * 12.55
        return elctric_charges
    elif monthly_units > 300 and monthly_units <= 400:
        elctric_charges = monthly_units * 17.38
        return elctric_charges
    elif monthly_units > 400 and monthly_units <= 500:
        elctric_charges = monthly_units * 18.84
        return elctric_charges
    elif monthly_units > 500 and monthly_units <= 600:
        elctric_charges = monthly_units * 19.76
        return elctric_charges
    elif monthly_units > 600 and monthly_units <= 700:
        elctric_charges = monthly_units * 20.40
        return elctric_charges
    else:
        elctric_charges = monthly_units * 23.87
        return elctric_charges
def general_sales_tax(gst_rate):   # function for GST calculation
    sales = electricity(monthly_units)
    gst_calculation = sales * gst_rate
    gst_value = round(gst_calculation, 2)
    return gst_value
def uniform_quarterly_adjustment():
    quarterly_adjustment = round(units_adjusted * rate_per_adjusted_unit, 2)
    return quarterly_adjustment
def elctricity_duty(monthly_units):
    elec_duty = electricity(monthly_units) * rate
    return elec_duty
def govt_charges():
    govt = general_sales_tax(gst_rate) + elctricity_duty(monthly_units)
    return govt
def ke_main():
    monthly_units = int(input("Enter monthly units: "))                     
    gst_rate = float(input("Enter GST rate: "))                               
    units_adjusted = float(input("Enter adjusted units: "))                    
    rate_per_adjusted_unit = float(input("Enter rate per adjusted units: ")) 
    rate = float(input("Enter Electricity duty rate: "))                      
    total_bill_amount = electricity(monthly_units) + uniform_quarterly_adjustment() + general_sales_tax(gst_rate) + elctricity_duty(monthly_units) + govt_charges()
    print("Electric Charges for the month is: \n", electricity(monthly_units))
    print("Uniform Quarterly Adjustment for the month is: \n", uniform_quarterly_adjustment())
    print("General Sales Tax for the month is: \n", general_sales_tax(gst_rate))
    print("Electricity Duty for the month is: \n", elctricity_duty(monthly_units))
    print("Government charges for the month is: \n", govt_charges())
    print("Amount Payable for the month is: \n", total_bill_amount)
ke_main()
```
